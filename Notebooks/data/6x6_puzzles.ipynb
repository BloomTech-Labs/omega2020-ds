```
import pandas as pd
from google.colab import files
```

```
# loading and looking at the data
df = pd.read_csv("uncleaned_sudokus_6x6.csv")
print(df.shape)
df.head()
```

```
# cleaning dataset
clean_df = df.copy()
clean_df.head()
```

```
# replace '0' with '.'
clean_df['puzzle'] = clean_df['puzzle'].replace(r'0', '.', regex = True)
clean_df.head()
```

```
# remove spaces
clean_df = clean_df.replace(r' ', '', regex = True)
clean_df.head()
```

```
# making a 'level' column
clean_df['level'] = 'Easy'
clean_df.head()
```

```
# rename 'puzzle' to 'sudoku'
clean_df.rename(columns = {'puzzle': 'sudoku'}, inplace = True)
clean_df.head()
```

```
# download the clean csv
clean_df.to_csv('6x6_puzzles.csv')
```
